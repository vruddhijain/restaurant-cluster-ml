```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings('ignore')

# Load the individual datasets
cust_data = pd.read_csv(r"D:\restoin\chefmozcuisine.csv")
cust_data1 = pd.read_csv(r"D:\restoin\rating_final.csv")
cust_data2 = pd.read_csv(r"D:\restoin\chefmozparking.csv")

# Merge the datasets based on the common attribute
merged_data = pd.merge(cust_data, cust_data1, on='placeID')
merged_data = pd.merge(merged_data, cust_data2, on='placeID')

# Perform one-hot encoding for the 'Rcuisine' column
encoder = OneHotEncoder(sparse_output=False)
encoded_cuisine = encoder.fit_transform(merged_data[['Rcuisine']])
encoded_cuisine_df = pd.DataFrame(encoded_cuisine, columns=encoder.get_feature_names_out(['Rcuisine']))
merged_data = pd.concat([merged_data, encoded_cuisine_df], axis=1)

# Perform one-hot encoding for the 'parking_lot' column
encoded_parking = encoder.fit_transform(merged_data[['parking_lot']])
encoded_parking_df = pd.DataFrame(encoded_parking, columns=encoder.get_feature_names_out(['parking_lot']))
merged_data = pd.concat([merged_data, encoded_parking_df], axis=1)

# Perform clustering on the merged dataset
attributes = ['rating', 'food_rating', 'service_rating']
attributes += list(encoded_cuisine_df.columns)  # Include the encoded cuisine columns
attributes += list(encoded_parking_df.columns)  # Include the encoded parking columns
X = merged_data[attributes]
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)
cluster_labels = kmeans.labels_
merged_data['cluster_label'] = cluster_labels
for cluster in range(num_clusters):
    cluster_data = merged_data[merged_data['cluster_label'] == cluster]
    print(f"Cluster {cluster + 1}:")
    print(cluster_data)
    print("\n")

# Save the merged dataset to a CSV file in the desired directory
output_path = r"D:\restoin\merged_dataset.csv"
merged_data.to_csv(output_path, index=False)
```
